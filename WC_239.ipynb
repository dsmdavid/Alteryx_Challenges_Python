## Brief

For this week's challenge, we will use Designer to solve/prove a classic verbal arithmetic problem. In this case, each letter in the equation "SEND + MORE = MONEY" will have a unique value between 0 and 9. Your job is to find the solution that makes the equation true.

```python
import pandas as pd
import itertools
import re
```

## Approach I want to follow:
1. All letters in "send + more = money" must have a value between 0-9.  
1. Create all permutations (10,8) and assign values. 
1. Validate those that make the equation true.

## 1. Create permutations:

```python
eq = "send + more = money"
words = ['send','more','money']
letters = list(set(list(re.sub('[^a-z]','',eq))))

values = list(range(0,10))
```

```python
combs = itertools.permutations(values,8)
```

```python
df = pd.DataFrame.from_records(combs,columns=letters )
```

```python
df.head()
```

## 2. Cast as string and create the words

```python
df = df.astype(str)
```

```python
for word in words:
    df[word] =''
    for letter in word:
        df[word] += df[letter]
```

## 3. Keep only those that make the equation true

```python
df['check'] = df['send'].astype(int) + df['more'].astype(int) == df['money'].astype(int)

df[df['check']].reset_index(drop=True)
```

```python
## Remove those that start with 0/

df['zero'] = (df['send'].apply(lambda x: x[0]=='0')) | (df['more'].apply(lambda x: x[0]=='0')) | (df['money'].apply(lambda x: x[0]=='0'))
```

```python
df[(df['check']) & (df['zero']==False)]
```
